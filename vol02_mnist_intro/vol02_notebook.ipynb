## 1. MNIST の概要

MNIST（Mixed National Institute of Standards and Technology database）は、  
手書き数字（0〜9）の画像データセットです。

- 画像サイズ：28×28 ピクセル  
- 色：グレースケール  
- 学習用：60,000 枚  
- テスト用：10,000 枚  

ディープラーニング入門で最も使われるデータセットです。

###MNISTをダウンロードして、画像データを1枚表示

```python
import torch
from torchvision import datasets, transforms
import matplotlib.pyplot as plt

# MNISTをダウンロード（初回のみ）
transform = transforms.ToTensor()
train_dataset = datasets.MNIST(root="./data", train=True, download=True, transform=transform)

print("学習データの枚数:", len(train_dataset))

# 1枚取り出して表示
image, label = train_dataset[0]

plt.imshow(image.squeeze(), cmap="gray")
plt.title(f"Label: {label}")
plt.axis("off")
plt.show()
```

###MNISTの画像データを表示

```python
import numpy as np

fig = plt.figure(figsize=(10, 4))

for i in range(20):
    img, lbl = train_dataset[i]
    ax = fig.add_subplot(2, 10, i + 1)
    ax.imshow(img.squeeze(), cmap="gray")
    ax.set_title(lbl)
    ax.axis("off")

plt.tight_layout()
plt.show()
```

## 2. 10クラス分類とは？

MNIST は **0〜9 の10種類の数字を分類するタスク** です。

モデルは画像を入力として受け取り、  
「これは 0 だと思う」「これは 7 だと思う」  
といった **10個のクラスの中から1つを選びます**。

この構造は他の分類問題にも応用できます。
